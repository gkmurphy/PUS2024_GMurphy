```python
#Central Tendancy of Philly Park Sizes
meanps = pparks['ACREAGE'].mean()
medianps = pparks['ACREAGE'].median()

print(f"The mean is {meanps} and the median is {medianps}")
```

**Conclusions:** The mean is considerably larger than the median suggesting that there are more smaller parks overall, but some very large parks that are skewing the overall average.

```python
# visualize a histogram of the park size.
# you can use the method of the ACREAGE column .hist()
# choose an appropriate number of bins with the argument bins=
# overplot on the histogram the mean and the median of the distribution
import matplotlib.pyplot as plt

ax = pparks["ACREAGE"].hist(bins=30)
ax.axvline(meanps, color='red', linewidth=2, label="Mean")
ax.axvline(medianps, color='orange', linewidth=2, label="Median")

ax.legend(); #to get the legend to render pass to axvline the argument label="..." with ... being whatever you want displayed
plt.xlabel('Park Size in Acres')
plt.ylabel('Number of Parks')
```

**Figure 1:** This figure shows the frequency and size of parks in Philadelphia as compared to the average and median size of parks as seen in red and orange, respectively. The mean is off of center due to the large amount of smaller parks and a few large park outliers.

```python
stdps = pparks['ACREAGE'].std()

print(f"The standard deviation is {stdps}")
```

# Analysis of Parks < 10 acres
calculate the mean and median and plot a histogram of the park sized with the mean and median indicated on the plot




```python
SmPparks = pparks[pparks['ACREAGE'] <= 10]
SmPparks.describe()

SmMeanps = SmPparks['ACREAGE'].mean()
SmMedianps = SmPparks['ACREAGE'].median()


ax = SmPparks["ACREAGE"].hist(bins=30)
ax.axvline(SmMeanps, color='red', linewidth=2, label="Mean")
ax.axvline(SmMedianps, color='orange', linewidth=2, label="Median")

ax.legend()
plt.xlabel('Park Size in Acres')
plt.ylabel('Number of Parks')
plt.title('Philadelphia Parks Smaller Than 10 Acres')
```

**Figure 2:** This figure shows the frequency and size of parks in Philadelphia that are less than 10 acres in size as compared to the average and median size of parks as seen in red and orange, respectively. The mean and median is closer to center than the larger dataset of parks, but the data is still skewed left due to the large amount of smaller parks.

# Group by Parent Park

```python
GrpPparks = pparks.groupby("PARENT_NAME")["ACREAGE"].sum()

GrpMeanps = GrpPparks.mean()
GrpMedianps = GrpPparks.median()

ax = SmPparks["ACREAGE"].hist(bins=30)
ax.axvline(GrpMeanps, color='red', linewidth=2, label="Mean")
ax.axvline(GrpMedianps, color='orange', linewidth=2, label="Median")

ax.legend()
plt.xlabel('Park Size in Acres')
plt.ylabel('Number of Parks')
plt.title('Philadelphia Parks Grouped by Their Parent Park')
```

**Figure 3:** This figure shows the frequency and size of parks in Philadelphia grouped by their park park in size as compared to the average and median size of parks as seen in red and orange, respectively. There is a large difference between the average and median showing a significant skew to the left with many of parent groups with low acreage.

The gap between the two metrics is very surprising - it would be interesting to bin parks by their size and parent groups to try and tease out what the difference is between the parks.
